```yml
- if cell with value 0 has at least one 1 from up, down, left or right side then you have to convert this change it to 'b';

- if cell has value 'b' its bin cell;

- Find clothest distance from every cell to bin cell;
```

````python
matrix = np.random.choice((0, 1), size=(100, 100))
````

````python
def is_bin_cell(x, y, matrix):
    try:
        up = matrix[x-1][y] if x > 0 else 0
    except:
        up = 0

    try:
        down = matrix[x+1][y] if x < len(matrix) else 0
    except:
        down = 0

    try:
        matrix[x][y-1] if y > 0 else 0
    except:
        down = 0

    try:
        left = matrix[x][y-1] if y > 0 else 0
    except:
        left = 0

    try:
        right = matrix[x][y+1] if y < len(matrix[0]) else 0
    except:
        right = 0
    
    if (matrix[x][y] == 0) and (1 in [up, down, right, left]):
        return True
    return False


def get_bin_cells(matrix):
    bindexes = []
    for i, row in enumerate(matrix):
        for j, _ in enumerate(row):
            if is_bin_cell(i, j, matrix):
                bindexes.append((i, j))
    return bindexes


def euclidean(i, j, matrix):
    p1, p2 = i
    d1, d2 = j
    return (((matrix[p1][0] - matrix[d1][0])**2) + ((matrix[p2][1] - matrix[d2][1])**2)) ** .5


def get_distances(matrix):
    bin_cells = get_bin_cells(matrix)
    distances = {}
    for i, row in enumerate(matrix):
        for j, _ in enumerate(row):
            distances[f'({i}, {j})'] = [euclidean((i, j), t, matrix) for t in bin_cells]

    return distances

````

````python
# driver code
data = get_distances(matrix)
index = [str(i) for i in data]
rows = [[coord] + i for coord, i in zip(index, data.values())]
pd.DataFrame(rows, columns=['Coord']+get_bin_cells(matrix))
````

```yml
Given an array of n strings with length m return true if there is one pair such that there is only one difference between the words, e.g, abcd and abce would yield true, but abcd and abdd would be false.
```

````python
def str_presence(a, b):
    a, b = list(set(a)), list(set(b))
    all_a = [i for i in a if i in b]
    all_b = [i for i in b if i in a]
    return (len(all_a) == len(b)) or (len(all_b) == len(a))

pd.DataFrame(
    [
        ['abcd', 'abce', str_presence("abcd", "abce")],
        ['abcd', 'abdd', str_presence("abcd", "abdd")]
    ],
    columns=['Input A', 'Input B', 'Output']
)
````

```yml
When a person who knows it meets any other person, they immediately share the story with them.

Initially, only person 1 knows the story. Given a list of meetings between people in a form of (person_1_id, person_2_id, timestamp) construct a list of the persons who will know the story at the very end.

Example:

[(1, 2, 100), (3,4, 200), (1,3, 300), (2,5, 400)], 1 // The events could be out of order.
Person 2 will learn the story at the moment 100, person 3 — at the moment 300,
person 5 — in the moment 400. Person 4 will never learn the story.
Output: [1, 2, 3, 5].

Eg2:

[(1, 2, 100), (2, 3, 100), (4, 5, 100)], 2
where the first parameter is array of the Persons meet at particular timestamp, second parameter is the PersonId who knows the story first.
Output: [1, 2, 3]
```

````python
def knows_the_story(people_timeline, initial):
    story = [initial]
    timeline = sorted(people_timeline, key=lambda t: t[2])

    for event in timeline:
        people_1, people_2, _ = event
        if people_1 in story and people_2 not in story:
            story.append(people_2)
        if people_2 in story and people_1 not in story:
            story.append(people_1)

    return story


# Driver code
sample_1 = [(1, 2, 100), (3,4, 200), (1,3, 300), (2,5, 400)]
sample_2 = [(1, 2, 100), (2, 3, 100), (4, 5, 100)]

pd.DataFrame(
    [
        [sample_1, 1, knows_the_story(sample_1, 1)],
        [sample_2, 2, knows_the_story(sample_2, 2)]
    ],
    columns=['Timeline', 'Initial  person', 'Output']
)
````

```yml
The minimum sum is 1 + 3 + 5 + 7 = 16 and the maximum sum is 3 + 5 + 7 + 9 = 24. The function prints:
```

````python
def minimax_sum(arr):
    # Write your code here
    hi = 0
    lo = sum(arr)
    handling = arr.copy()

    for _ in arr:
        value = sum(handling[:4])
        lo = value if value < lo else lo
        hi = value if value > hi else hi
        handling.append(handling.pop(0))

    return f'{lo} {hi}'

    
def minimax2(arr):
    hi = max(arr)
    lo = min(arr)

    hi_arr = arr.copy()
    lo_arr = arr.copy()
    hi_arr.pop(hi_arr.index(lo))
    lo_arr.pop(lo_arr.index(hi))

    return f'{sum(lo_arr)} {sum(hi_arr)}'


sample_1 = [1,3,5,7,9]
sample_2 = [5,8,5,0,10]
pd.DataFrame(
    [
        [sample_1, minimax_sum(sample_1), minimax2(sample_1)],
        [sample_2, minimax_sum(sample_2), minimax2(sample_2)]
    ],
    columns=['Input', 'Output O(n)', 'Output O(log n)']
)
````

```yml

Given a time in -hour AM/PM format, convert it to military (24-hour) time.

Note: - 12:00:00AM on a 12-hour clock is 00:00:00 on a 24-hour clock.
- 12:00:00PM on a 12-hour clock is 12:00:00 on a 24-hour clock.
```

````python
def time_conversion(s):
    handler = s[-2:]
    parts = s[:-2].split(':')
    switch = {
        '01': '13',
        '02': '14',
        '03': '15',
        '04': '16',
        '05': '17',
        '06': '18',
        '07': '19',
        '08': '20',
        '09': '21',
        '10': '22',
        '11': '23',
        '12': '12'
    }
    if handler.lower() == 'pm':
        parts[0] = switch[parts[0]]
    elif parts[0] == '12':
        parts[0] = '00'
    return ':'.join(parts)

# Driver code
sample_1 = '07:05:45PM'
sample_2 = '03:45:00PM'
sample_3 = '09:05:45AM'
sample_4 = '12:16:25PM'
sample_5 = '12:16:25AM'
pd.DataFrame(
    [
        [sample_1, time_conversion(sample_1)],
        [sample_2, time_conversion(sample_2)],
        [sample_3, time_conversion(sample_3)],
        [sample_4, time_conversion(sample_4)],
        [sample_5, time_conversion(sample_5)]
    ],
    columns=['Input', 'Output']
)
````

```yml
Given a list L of video names and their watch rates, write a function that will return the videos with the top 10 watch rates. Video names may appear more than once.

Example:

L = [(‘abc’, 10), (‘def’, 15), (‘ghi’, 10), (‘abc’, 12), …, (‘xyz’, 100)]

The function should return [‘xyz’, ‘abc’, …, ‘def’, ‘ghi’]
```

````python
def movie_ranking(L):
    rates = {}
    last = len(L) - 1
    for i in range(len(L)):
        # binary seaches the highest ratings for each unique movie
        head = L[i]
        tail = L[last]
        
        if not head[0] in rates:
            rates[head[0]] = head[1]
        else:
            rates[head[0]] = head[1] if head[1] > rates[head[0]] else rates[head[0]]

        if not tail[0] in rates:
            rates[tail[0]] = tail[1]
        else:
            rates[tail[0]] = tail[1] if tail[1] > rates[tail[0]] else rates[tail[0]]

    return sorted(list(rates.items()), key=lambda m: m[1], reverse=True)


random_names = [''.join(choice(list(letters[:26])) for _ in range(3)) for _ in range(5)]

sample_1 = [('abc', 10), ('def', 15), ('ghi', 10), ('abc', 12), ('xyz', 100)]
sample_2 = [('abc', 2), ('def', 15), ('ghi', 10), ('abc', 12), ('xyz', 100)]
sample_3 = [(choice(random_names), randint(1, 100)) for _ in range(5)]
sample_4 = [(choice(random_names), randint(1, 100)) for _ in range(5)]
sample_5 = [(choice(random_names), randint(1, 100)) for _ in range(5)]

pd.DataFrame(
    [
        [sample_1, movie_ranking(sample_1)],
        [sample_2, movie_ranking(sample_2)],
        [sample_3, movie_ranking(sample_3)],
        [sample_4, movie_ranking(sample_4)],
        [sample_5, movie_ranking(sample_5)]
    ],
    columns=['Input', 'Output']
)
````

```yml
A cord tree is a binary tree of strings.
A node in this tree can be a leaf node or an internal node.
An internal node has two children, a left child and a right child. It also has a length of all the children under it
A leaf nodes have a value and a length

                                #      InternalNode, 26
                                #      /              \   
                                #     /                \                         
                                #    /                  \
                                # Leaf(5, ABCDE)      InternalNode, 21
                                #                       /           \
                                #                      /             \
                                #                     /               \
                                #                    /                 \
                                #         Leaf(10, FGHIJKLMNO)     Leaf(11, PQRSTUVWXYZ)  

Q1: Define a Data Structure that represents a Cord tree.
Q2: Define a function that takes in a tree and an index and returns the character at that index.
```

````python
# PERSONAL NOTE: This is a very better implementation than the one i di in the interview.
# This one uses class properties to get the string value and the length
# On my interview i have implemented a fail DPS that sucks, how could i?
# Why do we think in the best ways to do something after we already did it?


class Leaf:
    def __init__(self, data):
        self.data = data
    
    @property
    def length(self):
        return len(self.data)

    def __str__(self):
        return 'leaf'

    def __len__(self):
        return len(self.data)


class Node:
    def __init__(self, left, right):
        self.left = left
        self.right = right

    @property
    def length(self):
        return len(self.left) + len(self.right)
    
    @property
    def data(self):
        return self.left.data + self.right.data

    def __len__(self):
        return len(self.left) + len(self.right)

    def __str__(self):
        return 'node'


class Tree:
    def __init__(self, root):
        self.root = root
        self.data = self.root.data

    def __str__(self):
        return 'tree'
        
    def __len__(self):
        return self.root.length
````

````python
# Driver code for tree
leaf_1 = Leaf('ABCDE')
leaf_2 = Leaf('FGHIJKLMNO')
leaf_3 = Leaf('PQRSTUVWXYZ')
node_1 = Node(leaf_2, leaf_3)
root = Node(leaf_1, node_1)
tree = Tree(root)

rows = [[str(tree), len(tree), str(tree.root), tree.data]]
nodes = [
    tree.root,
    tree.root.left,
    tree.root.right,
    tree.root.right.left,
    tree.root.right.right
]
for node in nodes:
    row = [str(node), node.length]
    if str(node) == 'node':
        row.append(f'{str(node.left)} | {node.right}')
    else:
        row.append(None)
    row.append(node.data)
    rows.append(row)

pd.DataFrame(
    rows,
    columns=['Type', 'Length', 'Childs', 'Data']
)
````
